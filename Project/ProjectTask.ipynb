```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
unrate=pd.read_csv("C:/Users/hp/Desktop/unrate.csv")
```

```python
unrate.head()
```

```python
unrate['DATE'] = pd.to_datetime(unrate['DATE'])
```

```python
unrate.head(12)
```

```python
plt.plot()
plt.show()
```

```python
#LinePlot
x=unrate.DATE.loc[:12]
y=unrate.VALUE.loc[:12]
plt.plot(x,y)
plt.xticks(rotation=90)
plt.xlabel('Month') 
plt.ylabel('Unemployment Rate')
plt.title('Monthly Unemployment Trends, 1948')
```

```python
#multiple plot for two years
fig = plt.figure(figsize=(12,5))
for i in range(2):
    ax = fig.add_subplot(2,1,i+1)
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    ax.plot(subset['DATE'], subset['VALUE'])
```

```python
#creating new coloumn
unrate['MONTH'] = unrate['DATE'].dt.month
```

```python
#subplots
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1) 
ax2 = fig.add_subplot(2, 1, 2)
ax1.plot(unrate['DATE'][:12], unrate['VALUE'][:12])
ax2.plot(unrate['DATE'][12:24], unrate['VALUE'][12:24])
```

```python
fig = plt.figure(figsize=(12, 5))
plt.plot(unrate[0:12]['MONTH'], unrate[0:12]['VALUE'], c='red',label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['VALUE'], c='blue', label='1949')
plt.plot(unrate[24:36]['MONTH'], unrate[24:36]['VALUE'], c='green', label='1950')
plt.legend(loc="upper left")
ax.set_title('Unemployment Trends for 3 years')
```

```python
reviews=pd.read_csv("C:/Users/hp/Desktop/fandango_scores.csv")
norm_reviews=reviews[['FILM','RT_user_norm','Metacritic_user_nom','IMDB_norm','Fandango_Ratingvalue','Fandango_Stars']].copy()
```

```python
#creating barplots
fig, ax = plt.subplots()
# Positions of the left sides of the 5 bars. [0.75, 1.75, 2.75, 3.75, 4.75]
from numpy import arange
bar_positions = arange(5) + 0.75

# Heights of the bars.  In our case, the average rating for the first movie in the dataset.
num_cols = ['RT_user_norm', 'Metacritic_user_nom', 'IMDB_norm', 'Fandango_Ratingvalue', 'Fandango_Stars']
bar_heights = norm_reviews[num_cols].iloc[0].values

ax.bar(bar_positions, bar_heights)
```
